```python
import google.generativeai as genai
from google.colab import userdata
import google.generativeai.types as glm

GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)

model = genai.GenerativeModel("gemini-2.0-flash-exp")

print("Welcome! Ask me anything. Type 'STOP' to exit.")
chat = model.start_chat()
system_prompt = "You are a helpful assistant who will first check if query should be flagged or not. If the query should not be flagged, you will answer user queries. Your creater is Aisha Batool."
count = 0
while True:
    s = input("User: ")
    if s.upper() == "STOP":
        print("Thanks for using our Custom LLM!")
        break
    response = chat.send_message("Check if the given string should be flagged: '"+ s+"'. If it should be flagged, say 'NOT SAFE CONTENT'")
    if response.text.strip()!='NOT SAFE CONTENT':
        response = chat.send_message("Okay, now please answer this question: "+ s)
        print("Gemini:", response.text)
    else:
        count+=1
        if count==3:
          print("You can not ask any further questions due to three violations!")
          break
        else:
          print("Your question violates our safety guidelines. You have", 3-count, "chance/s before you are blocked!")
```
